# **Matplotlib**

[Matplotlib](https://matplotlib.org/) is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
import matplotlib.pyplot as plt
import numpy as np
```

- Plot a line from position `(0, 0)` to position `(8, 250)`. Coordinates in `(x, y)`.

```python
x = np.array([0, 8])
y = np.array([0, 250])

plt.plot(x,y)
plt.show()
```

- Draw two points in the diagram, one at position `(1, 3)` and one in position `(8, 10)`.

```python
x = np.array([1,8])
y = np.array([3,10])

plt.plot(x , y ,'o')
plt.show()
```

- Draw a line in a diagram from position `(1, 3)` to `(2, 8)` then to `(6, 1)` and finally to position `(8, 10)`.

```python
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])

plt.plot(x,y)
plt.show()
```

## **Matplotlib Markers**

- Mark each point with a circle.

```python
y = np.array([3,8,1,10])
plt.plot(y,marker = 'o')
plt.show()
```

- Mark each point with a star.

```python
y = np.array([3,8,1,10])
plt.plot(y,marker='*')
plt.show()
```

## **Line Color**

```python
y = np.array([3,8,1,10])
plt.plot(y, marker='^', color='g') # triangle marker, green color
plt.show()
```

```python
y = np.array([3,8,1,10])
plt.plot(y, marker='p', color='red') # pentagon marker
plt.show()
```

### **Marker, linestyle, marker color(mc)**

```python
y = np.array([3,8,1,10])
plt.plot(y,'o:m') # marker, linestyle, marker color
plt.show()
```

## **Marker size**

`ms` is marker size.

```python
y = np.array([3,8,1,10])
plt.plot(y,marker = '*',ms = 25)
plt.show()
```

## **Marker colors**

There are two parameters available to colorize the matplotlib markers:

- Matplotlib Marker Edge Color (mec).

- Matplotlib Marker Face Color (mfc).

```python
y = np.array([3,8,1,10])
plt.plot(y ,'o:r', ms = 10, mec='c')
plt.show()
```

Marker face color `mfc`

```python
y = np.array([3,8,1,10])
plt.plot(y, 'o:r', ms = 10, mec ='y', mfc='b')
plt.show()
```

Marker edge width `mew`

```python
y = np.array([3,8,1,10])
plt.plot(y, 'o:r', ms = 10, mec ='y', mfc='k', mew=2)
plt.show()
```
